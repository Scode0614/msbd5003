```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.offline as py
```

```python
data = pd.read_csv("time_series_19-covid-Confirmed_archived_0325.csv")
data.head()
```

```python
China_data = data[data['Country/Region'] == 'China']
China_data.head()
```

```python
py.init_notebook_mode(connected=True)
fig = go.Figure()
for index, row in China_data.iterrows():
    fig.add_trace(go.Scatter(x = row.index[4:], y = list(row)[4:], name = row[0], line=dict(width=4)))
fig.update_layout(title='Province Confimred Case Number',
                   xaxis_title='Date',
                   yaxis_title='Cases')
py.iplot(fig)
```

```python
USA_data = data[data['Country/Region'] == 'US']
```

```python
fig = go.Figure()
for index, row in USA_data.iterrows():
    fig.add_trace(go.Scatter(x = row.index[4:], y = list(row)[4:], name = row[0], line=dict(width=4)))
fig.update_layout(title='State in USA Confimred Case Number',
                   xaxis_title='Date',
                   yaxis_title='Cases')
py.iplot(fig)
```

```python
from pyecharts.charts import Geo
from pyecharts import options as opts
from pyecharts.globals import GeoType

def geo_base(df):
    g = Geo()
    g.add_schema(maptype="world")
    for i in range(len(df)): 
        g.add_coordinate(df.iloc[i]['Province/State'],df.iloc[i]['Long'],df.iloc[i]['Lat'])
    data_pair=[list(z) for z in zip(df['Province/State'], df['3/23/20'])] #新版的数值
    g.add('confirmed number', data_pair,symbol_size=8)
    g.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    g.set_global_opts(
            visualmap_opts=opts.VisualMapOpts(min_= 0, max_= 1000),
            title_opts=opts.TitleOpts(title="confirmed heat map")
        )
    return g
g = geo_base(data)
g.render_notebook() 
```
